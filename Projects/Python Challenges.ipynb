```python
# ************************Python Challenges***********************************
#  Challenge: Student Marks Analyzer
#
# Create a Python program that allows a user to input student names along with their marks and then calculates useful statistics.
# Your program should:
# 1. Let the user input multiple students with their marks (name + integer score).
# 2. After input is complete, display:
#    - Average marks
#    - Highest marks and student(s) who scored it
#    - Lowest marks and student(s) who scored it
#    - Total number of students
#
# Bonus:
# - Allow the user to enter all data first, then view the report
# - Format output clearly in a report-style layout
# - Prevent duplicate student names

info = {}

while True:
    name = input("Enter student name: ")

    # Prevent duplicate names
    if name in info:
        print("Student already exists! Try a different name.")
        continue

    marks = int(input("Enter marks: "))
    info[name] = marks

    choice = input("Add another student? (y/n): ").lower()
    if choice != 'y':
        break

# Calculation
total = sum(info.values())
length = len(info)
avg = total / length

# Find highest and lowest marks
highest = max(info.values())
lowest = min(info.values())

# Find students  name with highest & lowest marks
highest_students = [name for name, marks in info.items() if marks == highest]
lowest_students = [name for name, marks in info.items() if marks == lowest]

print("\n" + "="*30)
print("      STUDENT REPORT")
print("="*30)

print(f"Total Students     : {length}")
print(f"Average Marks      : {avg:.2f}")

print(f"Topper(s)          : {', '.join(highest_students).title()}")
print(f"Highest Marks      : {highest}")

print(f"Lowest Scorer(s)   : {', '.join(lowest_students).title()}")
print(f"Lowest Marks       : {lowest}")

print("="*30)
```
