```python
# 2. Write a Python program to read first n lines of a file.
n = 3
try:
    with open("sample_file1.txt", "r") as file:
        for i in range(n):
            line = file.readline()
            if not line:
                break
            print(line, end="")

except FileNotFoundError:
    print("File not found")
```

```python
# 7. Write a python program to find the longest words.

try:
    with open("sample_file1.txt", "r") as file:
        words = file.read().split()
        max_len = max(len(word) for word in words)

        longest_words = [word for word in words if len(word) == max_len]

        print("Longest words:", longest_words)

except FileNotFoundError:
    print("File not found")
```

```python
# 9. Write a Python program to count the frequency of words in a file.

try:
    freq = {}
    with open("sample_file1.txt", "r") as file:
        for line in file.readlines():
            for word in line.split():
                if word not in freq:
                    freq[word] = 1
                else:
                    freq[word] += 1
    print(freq)

except FileNotFoundError:
    print("File not found")
```

```python
# 12. Write a Python program to copy the contents of a file to another file .

try:
    with open("sample_file1.txt", "r") as source, open("sample_file2.txt", "w") as destination:
        for line in source:
            destination.write(line)

    with open("sample_file2.txt", "r") as file:
        print(file.read())

except Exception as e:
    print(e.args)
```

```python
# 14. Write a Python program to assess if a file is closed or not.

try:
    file = open("sample_file1.txt", "r")
    print("Is file closed?", file.closed)   # False (file is open)

    file.close()

    print("Is file closed?", file.closed)   # True (file is closed)

except FileNotFoundError:
    print("File not found")
```
