```python
# 1. Write a Python program to create a class representing a Circle. Include methods to calculate its area and perimeter.
import math
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        if self.radius <= 0:
            return None
        return math.pi * self.radius ** 2

    def perimeter(self):
        if self.radius <= 0:
            return None
        return 2 * math.pi * self.radius


c1 = Circle(7)

if c1.area() is not None:
    print("Area is:", c1.area())
else:
    print("Invalid input")

if c1.perimeter() is not None:
    print("Perimeter is:", c1.perimeter())
else:
    print("Invalid input")
```

```python
# 4. Write a Python program to create a class that represents a shape. Include methods to calculate its area and perimeter. Implement subclasses for different shapes like circle, triangle, and square.
import math

class Shape:
    def circle(self,radius):
        area = math.pi * radius ** 2
        perimeter = 2 * math.pi * radius
        return "area:-",area,"perimeter:-",perimeter

    def triangle(self, base, side1, side2, height):
        area = 0.5 * base * height
        perimeter = base + side1 + side2
        return "area:-",area,"perimeter:-",perimeter

    def square(self, side):
        area = side ** 2
        perimeter = 2 * side
        return "area:-",area,"perimeter:-",perimeter

s1 = Shape()
print("square:-",s1.square(3))
print("triangle:-",s1.triangle(3,4,5,6))
```

```python
# 8. Write a Python program to create a class representing a shopping cart. Include methods for adding and removing items, and calculating the total price.

class ShoppingCart:
    def __init__(self):
        self.products = {}

    def add_items(self, items, price):
        if items not in self.products:
            self.products[items] = price
        else:
            print("Item already exists")

    def remove_items(self, items):
        if items not in self.products:
            print("Item not exists")
        else:
            del self.products[items]

    def total_price(self):
        total = 0
        print("Products in cart:")
        for key, values in self.products.items():
            print(key, "->", values)
            total += values
        print("Total price:", total)

s = ShoppingCart()
s.add_items("apple", 100)
s.add_items("banana", 200)
s.add_items("Oil", 300)
s.add_items("water", 100)
s.add_items("rice", 300)

s.remove_items("banana")

s.total_price()
```

```python
# 11. Write a Python program to create a class representing a bank. Include methods for managing customer accounts and transactions.
# 	Create a Python class called BankAccount which represents a bank account, having as attributes: accountNumber (numeric type), name (name of the account owner as string type), balance.
# 	Create a constructor with parameters: accountNumber, name, balance.
# 	Create a Deposit() method which manages the deposit actions.
# 	Create a Withdrawal() method which manages withdrawals actions.
# 	Create an bankFees() method to apply the bank fees with a percentage of 5% of the balance account.
# 	Create a display() method to display account details. Give the complete code for the BankAccount class.

class BankAccount:
    def __init__(self, accountNumber, name, balance):
        self.accountNumber = accountNumber
        self.name = name
        self.balance = balance
        self.transactions = []

    def deposit(self, amount):
        if amount < 0:
            print("Invalid amount")
        else:
            self.balance += amount
            self.transactions.append(("deposit", amount))
            print("Deposited:", amount)

    def withdraw(self, amount):
        if amount < 0:
            print("Invalid amount")
        else:
            self.balance -= amount
            self.transactions.append(("withdraw", amount))
            print("Withdrawn:", amount)

    def bankFees(self):
        if self.balance == 0:
            print("No Bank fees applied.")
        else:
            fees = self.balance * 0.05
            print("Bank Fees:", fees)

    def display(self):
        print("Account Number:", self.accountNumber)
        print("Name:", self.name)
        print("Balance:", self.balance)
        print("Transactions:", self.transactions)

c1 = BankAccount(234567, "Parag", 50000)
c2 = BankAccount(456767, "Prasad", 40000)

c1.deposit(10000)
c1.withdraw(15000)
c1.bankFees()
c1.display()
```

```python
# 13. TechWorld, a technology training center, wants to allocate courses for instructors. An instructor is identified by name, technology skills, experience and average feedback. An instructor is allocated a course, if he/she satisfies the below two conditions:
#
# 	eligibility criteria:
# 	if experience is more than 3 years, average feedback should be 4.5 or more
# 	if experience is 3 years or less, average feedback should be 4 or more
# 	he/she should possess the technology skill for the course
# 	Identify the class name and attributes to represent instructors. Write a Python program to implement the class chosen with its attributes and methods.
#
# 	Note:
# 	Consider all instance variables to be private and methods to be public.
# 	An instructor may have multiple technology skills, so consider instance variable, technology_skill to be a list.
# 	check_eligibility(): Return true if eligibility criteria is satisfied by the instructor. Else, return false
# 	allocate_course(technology): Return true if the course which requires the given technology can be allocated to the instructor. Else, return false.
# 	Represent a few objects of the class, initialize instance variables using setter methods, invoke appropriate methods and test your program.

class Instructor:
    def __init__(self):
        self.__name = None
        self.__tech_skills = []
        self.__experience = 0
        self.__avg_feedback = 0

    def setter(self, name, tech_skills, experience, avg_feedback):
        self.__name = name
        self.__tech_skills = tech_skills
        self.__experience = experience
        self.__avg_feedback = avg_feedback

    def getter(self):
        return self.__name

    def check_eligibility(self):
        if self.__experience > 3 and self.__avg_feedback > 4.5:
            return True
        elif self.__experience <= 3 and self.__avg_feedback >= 4:
            return True
        else:
            return False

    def allocate_course(self, technology):
        if self.check_eligibility():
            if technology in self.__tech_skills:
                return True
        return False


instructors = []

# Adding multiple instructors
while True:
    ins = Instructor()
    name = input("Enter Instructor Name: ")
    tech_skills = input("Enter skills (comma separated): ").split(",")
    tech_skills = [skill.strip() for skill in tech_skills]
    experience = int(input("Enter Experience: "))
    avg_feedback = float(input("Enter Feedback (0-5): "))

    ins.setter(name, tech_skills, experience, avg_feedback)
    instructors.append(ins)

    ch = input("Add more instructors (y/n)? ").lower()
    if ch != 'y':
        break

# Requirement
tech = input("Enter required technology: ")

found = False

for ins in instructors:
    if ins.allocate_course(tech):
        print(f"{ins.getter()} can teach {tech}")
        found = True

if not found:
    print("No suitable instructor found")

```

```python
# 18. Write a Python class Employee with attributes like emp_id, emp_name, emp_salary, and emp_department and methods like calculate_emp_salary, emp_assign_department, and print_employee_details.
#
# 	Sample Employee Data:
# 	"ADAMS", "E7876", 50000, "ACCOUNTING"
# 	"JONES", "E7499", 45000, "RESEARCH"
# 	"MARTIN", "E7900", 50000, "SALES"
# 	"SMITH", "E7698", 55000, "OPERATIONS"
#
# 	Use 'assign_department' method to change the department of an employee.
# 	Use 'print_employee_details' method to print the details of an employee.
# 	Use 'calculate_emp_salary' method takes two arguments: salary and hours_worked, which is the number of hours worked by the employee. If the number of hours worked is more than 50, the method computes overtime and adds it to the salary. Overtime is calculated as following formula:
# 	overtime = hours_worked - 50
# 	Overtime amount = (overtime * (salary / 50))
class Employee:
    def __init__(self, emp_name, emp_id, emp_salary, emp_department):
        self.emp_id = emp_id
        self.emp_name = emp_name
        self.emp_salary = emp_salary
        self.emp_department = emp_department
        self.overtime = 0

    def calculate_emp_salary(self, hours_worked):
        salary = self.emp_salary
        if hours_worked > 50:
            overtime = hours_worked - 50
            overtime_amount = overtime * (salary / 50)
            total_salary = salary + overtime_amount
        else:
            overtime = 0
            total_salary = salary

        print(f"{self.emp_name}'s salary:", total_salary)

    def emp_assign_department(self, department):
        if self.emp_department.lower() == department.lower():
            print("Employee already assigned to department.")
        else:
            self.emp_department = department.upper()
            print(f"Employee assigned to department: {self.emp_department}")

    def print_employee_details(self):
        print("\nEmployee Details")
        print("Employee name:", self.emp_name)
        print("Employee ID:", self.emp_id)
        print("Employee Salary:", self.emp_salary)
        print("Overtime:", self.overtime,"Hrs")
        print("Employee Department:", self.emp_department)

e1 = Employee("ADAMS", "E7876", 50000, "ACCOUNTING")
e2 = Employee("JONES", "E7499", 45000, "RESEARCH")
e3 = Employee("MARTIN", "E7900", 50000, "SALES")
e4 = Employee("SMITH", "E7698", 55000, "OPERATIONS")

e1.calculate_emp_salary(55)

e3.emp_assign_department("sales")

e1.print_employee_details()
e3.print_employee_details()

```

```python
# 19. Write a Python class Restaurant with attributes like menu_items, book_table, and customer_orders, and methods like add_item_to_menu, book_tables, and customer_order.
# 	Perform the following tasks now:
# 	Now add items to the menu.
# 	Make table reservations.
# 	Take customer orders.
# 	Print the menu.
# 	Print table reservations.
# 	Print customer orders.
# 	Note: Use dictionaries and lists to store the data.

class Restaurant:
    def __init__(self):
        self.menu_items = {}
        self.book_table = []
        self.customer_orders = {}

    def add_item_to_menu(self, item, price):
        if item not in self.menu_items:
            self.menu_items[item] = price
        else:
            print(f"{item} already exists in menu.")

    def book_tables(self, table_no):
        if table_no not in self.book_table and table_no <= 30:
            self.book_table.append(table_no)
            print(f"Table {table_no} booked.")
        else:
            print(f"Table {table_no} not available.")

    def menu_card(self):
        print("\n------ MENU ------")
        for item, price in self.menu_items.items():
            print(f"{item} : Rs.{price}")

    def customer_order(self, table_no, item, qty):
        if item not in self.menu_items:
            print(f"{item} not available.")
            return

        if table_no not in self.customer_orders:
            self.customer_orders[table_no] = {}

        if item in self.customer_orders[table_no]:
            self.customer_orders[table_no][item] += qty
        else:
            self.customer_orders[table_no][item] = qty

    def print_table_reservations(self):
        print("\n--- Reserved Tables ---")
        print(self.book_table)

    def print_customer_orders(self):
        print("\n--- Customer Orders ---")
        for table, orders in self.customer_orders.items():
            print(f"Table {table}: {orders}")

    def bill(self, table_no):
        print("\n------ BILL ------")
        print(f"Table No: {table_no}")

        total = 0
        orders = self.customer_orders.get(table_no, {})

        for item, qty in orders.items():
            price = self.menu_items[item]
            print(f"{item} : {qty} x Rs.{price}")
            total += qty * price

        print("Total =", total)
r = Restaurant()

r.add_item_to_menu("Mix Veg", 250)
r.add_item_to_menu("Paneer Butter Masala", 320)
r.add_item_to_menu("Dal Tadka", 180)
r.add_item_to_menu("Veg Biryani", 220)
r.add_item_to_menu("Jeera Rice", 150)
r.add_item_to_menu("Butter Naan", 50)
r.add_item_to_menu("Masala Dosa", 120)
r.add_item_to_menu("Chole Bhature", 140)
r.add_item_to_menu("Veg Fried Rice", 200)
r.add_item_to_menu("Hakka Noodles", 190)
r.add_item_to_menu("Gulab Jamun", 15)

r.book_tables(12)

r.menu_card()

r.customer_order(12,"Paneer Butter Masala", 1)
r.customer_order(12,"Dal Tadka", 1)
r.customer_order(12,"Butter Naan", 2)
r.customer_order(12,"Jeera Rice", 1)
r.customer_order(12,"Gulab Jamun", 4)

# Print all
r.print_table_reservations()
r.print_customer_orders()

# Bill
r.bill(12)
```

```python
# 20. Write a Python class BankAccount with attributes like account_number, balance, date_of_opening and customer_name, and methods like deposit, withdraw, and check_balance.
from datetime import date
class BankAccount:
    def __init__(self, account_no, date_of_opening, name, balance):
        self.account_no = account_no
        self.date_of_opening = date_of_opening
        self.name = name
        self.balance = balance
        self.transactions = []

    def deposit(self, amount):
        if amount <= 0:
            print("Invalid amount.")
        else:
            self.balance += amount
            self.transactions.append(f"{amount} Deposit")
            print(f"{amount} deposited successfully.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance.")
        else:
            self.balance -= amount
            self.transactions.append(f"{amount} Withdraw")
            print(f"{amount} withdrawn successfully.")

    def check_balance(self):
        print(f"""Name:- {self.name}
Account No:- {self.account_no}
Balance:- {self.balance}
Transactions:- {self.transactions}""")

c1= BankAccount(23454, date(2024, 4, 2), "Parag", 50000)
c1.deposit(15000)
c1.withdraw(30000)
c1.check_balance()
```

```python
# 21. Write a Python class Inventory with attributes like item_id, item_name, stock_count, and price, and methods like add_item, update_item, and check_item_details.

class Inventory:
    def __init__(self):
        self.inventory = {}

    def add_item(self, item_id, item_name, stock_count, price):
        if item_name in self.inventory:
            self.inventory[item_name]["stock"] += stock_count
        else:
            self.inventory[item_name] = {
                "id": item_id,
                "stock": stock_count,
                "price": price
            }

    def update_item(self, item_name, stock_count):
        if item_name in self.inventory:
            self.inventory[item_name]["stock"] += stock_count
        else:
            print("Item not found add the item first.")

    def check_item_details(self):
        for key, value in self.inventory.items():
            print(f"{key}: {value['stock']} x {value['price']}")


inv = Inventory()

inv.add_item("E101", "Laptop", 5, 50000)
inv.add_item("E102", "Mouse", 10, 500)

inv.update_item("Laptop", 2)

inv.check_item_details()
```
