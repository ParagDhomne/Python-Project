```python
# 1. Write a Python program to handle a ZeroDivisionError exception when dividing a number by zero.

try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))
    result = num1 / num2
    print(result)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ValueError:
    print("Invalid input")
```

```python
# 2. Write a Python program that prompts the user to input an integer and raises a ValueError exception if the input is not a valid integer.

try:
    integer = int(input("Enter integer: "))
    print("Integer you have entered is:-", integer)

except ValueError:
    print("You have to enter an integer")
```

```python
# 3. Write a Python program that opens a file and handles a FileNotFoundError exception if the file does not exist.

try:
    with open("File.txt","r") as file:
        print(file.read())

except FileNotFoundError:
    print("File not found")
```

```python
# 4. Write a Python program that prompts the user to input two numbers and raises a TypeError exception if the inputs are not numerical.

try:
    a = float(input("Enter a number: "))
    b = input("Enter another number: ")
    c = a + b
    print(f"Addition of {a} and {b} is {c}")

except TypeError:
    print("One Number is not converted to number")
```

```python
# 5. Write a Python program that opens a file and handles a PermissionError exception if there is a permission issue.

try:
    filename = input("Enter file name: ")

    with open(filename, "r") as file:
        print(file.read())

except PermissionError:
    print("Permission denied")

except FileNotFoundError:
    print("File does not exist")

except Exception as e:
    print("Some other error:", e)
```

```python
# 6. Write a Python program that executes an operation on a list and handles an IndexError exception if the index is out of range.
my_list = [10, 20, 30, 40]

try:
    index = int(input("Enter index: "))

    if index < 0 or index >= len(my_list):
        raise IndexError

    print("Element:", my_list[index])

except IndexError:
    print("Invalid index position")

except ValueError:
    print("Enter a valid integer")
```

```python
# 7. Write a Python program that executes division and handles an ArithmeticError exception if there is an arithmetic error.

try:
    num1 = int(input("Enter first number: "))
    num2 = int(input("Enter second number: "))

    result = num1 / num2
    print("Result:", result)

except ZeroDivisionError:
    print("Cannot divide by zero")

except ArithmeticError:
    print("Some arithmetic error occurred")

except ValueError:
    print("Invalid input")
```
